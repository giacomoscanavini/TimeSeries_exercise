# LIGO data analysis for event GW150914

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import mlab # same function as plt.psd, but without the plot
import scipy.signal as sig
```

```python
# https://www.gw-openscience.org/eventapi/html/GWTC-1-confident/GW150914/v3/H-H1_GWOSC_16KHZ_R1-1126259447-32.txt.gz (Hanford data)
# https://www.gw-openscience.org/eventapi/html/GWTC-1-confident/GW150914/v3/L-L1_GWOSC_16KHZ_R1-1126259447-32.txt.gz (Livingston data)
Fs = 16384 # Sampling rate for data, Hz

# Skip 3 header rows
s_Hanford = np.loadtxt("data/H-H1_GWOSC_16KHZ_R1-1126259447-32.txt", skiprows=3)
s_Livingston = np.loadtxt("data/L-L1_GWOSC_16KHZ_R1-1126259447-32.txt", skiprows=3)

N = len(s_Hanford) # Number of points in the time stream
t = np.arange(N)/Fs # Array of time values for the sampling rate above, s
```

```python
## To ensure we've properly loaded the data, we will make a quick plot:
fig=plt.figure()
plt.plot(t, s_Hanford, label='Hanford')
plt.plot(t, s_Livingston, 'r', label="Livinston")
fig.set_size_inches(14,3)
plt.xlabel("Time (s)")
plt.ylabel("Strain, $\delta L/L$")
plt.legend()
plt.show()
```

The data above looks reasonable -- has the expected time of 32 s listed on the website, and total RMS strains ~1e-19 to ~1e-18. We will need to substantially improve the noise to see the ~1e-21 stain merger events.

```python
def plot_ASD(t, n, use_window=False, make_plot = True, NFFT="", label=""):
    """ Function to return and plot the amplitude spectral density (ASD)
      for a given vector n, as a function of time t

      Parameters:
        t - array, time vector (time of noise n in increasing order), assumed to be in seconds
        n - array of same length as t. noise values at time t, with some "units"
        use_window - boolean value to apply a Hanning window before calculating the ASD (default=False)
        make_plot - boolean value of whether to plot the ASD (default=True)
        NFFT - number of points to use for calculating the ASD. If this is specified (and not empty) 
               then Welch's method is used with this number of points
        label - String to label curve when plotting, if make_plot is True

      Returns:
        freqs - frequencies at which the ASD is calculated, Hz
        ASD - Amplitude spectral density, "units"/sqrt(Hz)
    """
    if(len(n) != len(t)):
        print("Error: time vector and noise vector must have same length")
        return 
    
    T = t[-1]-t[0] # find time between first and last samples
    N = len(n) # total number of samples
    Fs = N/T # sampling rate, Hz
    
    if(NFFT): 
        #use the built-in Welch's method function, at lowered frequency resolution
        PSD, freqs = mlab.psd( n, Fs=Fs, NFFT = NFFT )
    else: 
        # explicitly use rfft over the full vector for maximum frequency resolution
        window = sig.windows.hann(N) # Hanning window for calculating PSD, if desired
        #assume real sequence and use real fft
        if(use_window): g = np.fft.rfft(n*window) 
        else: g = np.fft.rfft(n)
        norm = 2*T/N**2 #norm for single sided discrete PSD in numpy convention
        PSD = norm*(np.abs(g))**2
        freqs = np.fft.rfftfreq(N, 1/Fs) # helper function to get frequency vector
    
    ASD = np.sqrt(PSD)
    if(make_plot):
        plt.loglog( freqs, ASD, label=label, alpha=0.75)
        plt.xlabel("Frequency [Hz]")
        plt.xlim([20, 7e3]) # Limits to match LIGO plot
        plt.ylim([1e-24, 1e-21]) # Limits to match LIGO plot
    
    return freqs, ASD
```

```python
## Make a plot of the ASD from the site at the highest frequency resolution 
## possible: 1/(32 s) = 31 mHz using the standard numpy rfft function:

plt.figure()
f_Hanford, asd_Hanford = plot_ASD(t, s_Hanford, label="Hanford")
f_Living, asd_Living = plot_ASD(t, s_Livingston, label="Livingston")
plt.ylabel("Strain noise, 1/$\sqrt{\mathrm{Hz}}$")
plt.legend()
plt.show()
```

So this looks somewhat like the expected LIGO spectrum, but there are some issues:


1.   It is very noisy between neighboring frequencies
2.   On the low frequency side, larger noise is seen than expected, especially for the Livingston site

We will fix those by windowing (to remove the excess low frequency noise artifacts) and by using Welch's method (to average down the noise on the ASD at the cost of frequency resolution) in the following.



```python
## Now apply Welch's method, for both sites:
NFFT = 2**14 # number of points per chunk (should be a power of two)

fig=plt.figure()
plt.subplot(1,2,1) ## first for Hanford
f_Hanford, asd_Hanford = plot_ASD(t, s_Hanford, label="Hanford")
f_Hanford_welch, asd_Hanford_welch = plot_ASD(t, s_Hanford, label="Hanford (Welch's meth.)", NFFT=NFFT)
plt.ylabel("Strain noise, 1/$\sqrt{\mathrm{Hz}}$")
plt.legend()

plt.subplot(1,2,2) ## Now for Livingston
f_Living, asd_Living = plot_ASD(t, s_Livingston, label="Livingston")
f_Living_welch, asd_Living_welch = plot_ASD(t, s_Livingston, label="Livingston (Welch's meth.)", NFFT=NFFT)
plt.ylabel("Strain noise, 1/$\sqrt{\mathrm{Hz}}$")
plt.legend()

fig.set_size_inches(13,4)
plt.show()
```

So Welch's method gives something that's much less noise, and also makes clear the excess noise for the un-windowed ASD, especially apparent at low frequency in the Livingston data. To remove this excess noise, we can apply a Hanning window as suggested in the write up:

```python
## Compare Welch's method to the version where we use rfft and apply a windowing
## function manually

NFFT = 2**14 # number of points per chunk (should be a power of two)

fig=plt.figure()
plt.subplot(1,2,1) ## first for Hanford
f_Hanford, asd_Hanford = plot_ASD(t, s_Hanford, label="Hanford", use_window=True) #now using a Hanning window
f_Hanford_welch, asd_Hanford_welch = plot_ASD(t, s_Hanford, label="Hanford (Welch's meth.)", NFFT=NFFT)
plt.ylabel("Strain noise, 1/$\sqrt{\mathrm{Hz}}$")
plt.legend()

plt.subplot(1,2,2) ## Now for Livingston
f_Living, asd_Living = plot_ASD(t, s_Livingston, label="Livingston", use_window=True) #now using a Hanning window
f_Living_welch, asd_Living_welch = plot_ASD(t, s_Livingston, label="Livingston (Welch's meth.)", NFFT=NFFT)
plt.ylabel("Strain noise, 1/$\sqrt{\mathrm{Hz}}$")
plt.legend()

fig.set_size_inches(13,4)
plt.show()
```

So the window fixes the issue with the excess low frequency noise using the rfft function naively. At this point the two methods agree (note that on the log scale, the downward noise fluctuations look asymmetric compared to the upward). Welch's method gives us lower noise, but if we zoom in on the noise lines we can see it has poorer resolution, e.g.:

```python
## example of zoom on linear scale (not needed for solution, but to illustrate some of the differences)
fig=plt.figure()
f_Hanford, asd_Hanford = plot_ASD(t, s_Hanford, label="Hanford", use_window=True) #now using a Hanning window
f_Hanford_welch, asd_Hanford_welch = plot_ASD(t, s_Hanford, label="Hanford (Welch's meth.)", NFFT=NFFT)
plt.ylabel("Strain noise, 1/$\sqrt{\mathrm{Hz}}$")
plt.gca().set_xscale('linear')
plt.xlim(20,100)
plt.legend()
plt.show()
```

```python
def apply_filters_fft(input_waveform, sampling_freq, band_list):
    """ Function to filter a waveform by removing frequencies
      within the given bands using an FFT based filter

      Parameters:
        input_waveform - waveform to filter
        sampling_freq - Sampling frequency, Hz
        band_list - array of start/stop frequencies, the FFT of the waveform
                    is zeroed in the range between each start/stop freq

      Returns:
        filtered_waveform - time stream of filtered waveform
    """  

    ## Assume real valued sequence and take fft of waveform
    g = np.fft.rfft(input_waveform)
    freqs  = np.fft.rfftfreq(len(input_waveform), 1/sampling_freq)

    ## loop over the filter bands and make a boolean array of frequencies to remove
    freqs_to_zero = freqs<0 
    for current_band in band_list:
        freqs_to_zero = np.logical_or(freqs_to_zero, ( (freqs>=current_band[0]) & (freqs<=current_band[1]) ))

    g[freqs_to_zero] = 0 ## zero out noisy frequencies
    filtered_waveform =  np.fft.irfft(g) ## inverse transform back to time domain

    return filtered_waveform
```

```python
## first make a zoomed in plot of the ASDs to see what frequencies we should remove
fig=plt.figure()
f_Hanford_welch, asd_Hanford_welch = plot_ASD(t, s_Hanford, label="Hanford", NFFT=2**16)
f_Living_welch, asd_Living_welch = plot_ASD(t, s_Livingston, label="Livingston", NFFT=2**16)
plt.gca().set_xscale('linear')
plt.gca().set_yscale('linear')
plt.ylabel("Strain noise, 1/$\sqrt{\mathrm{Hz}}$")
plt.xlim(20,500)
plt.ylim(0,1e-22)
fig.set_size_inches(15,4)

## find the bad frequencies to zero out
thresh = 0.25e-22
bad_freqs = (asd_Hanford_welch>thresh) | (asd_Living_welch>thresh)

min_freq, max_freq = 50, 400 ## reduce range to just get lowest noise part
bad_freqs = bad_freqs | (f_Hanford_welch < min_freq) | (f_Hanford_welch > max_freq)

## plot the bad frequency ranges
plt.fill_between(f_Hanford_welch, np.zeros_like(bad_freqs), bad_freqs, color='magenta', alpha=0.3)
plt.show()
```

```python
## now make a list of the ranges for the bad frequencies (this could also be done 
## in this case by hand from the plot above, with coarser selections)

bad_starts = np.argwhere(bad_freqs & (~np.roll(bad_freqs,1))) # indices for which this frequency is bad, but the previous one was good
bad_ends = np.argwhere(bad_freqs & (~np.roll(bad_freqs,-1))) # indices for which this frequency is bad, but the next one is good

bad_starts = np.insert(bad_starts,0,0) #insert zero at the beginning of the start array
bad_ends = np.append(bad_ends,-1) # add -1 to the end of the final block of frequencies

f=f_Hanford_welch
filter_list = np.vstack( (f[bad_starts], f[bad_ends]) ).T # make n x 2 array of bad intervals to pass to apply_filters_fft

print("Frequency ranges to remove:")
print(filter_list)
```

```python
## now we can make the ASD again to double check the filter is working as expected
s_Hanford_filt = apply_filters_fft(s_Hanford, Fs, filter_list)
s_Livingston_filt = apply_filters_fft(s_Livingston, Fs, filter_list)

fig=plt.figure()

#Hanford
plt.subplot(1,2,1)
f, s = plot_ASD(t, s_Hanford, NFFT=2**14, label='Before filter')
f, s = plot_ASD(t, s_Hanford_filt, NFFT=2**14, label='After filter')
plt.ylabel("Strain noise, 1/$\sqrt{\mathrm{Hz}}$")
plt.legend()

# Livingston
plt.subplot(1,2,2)
f, s = plot_ASD(t, s_Livingston, NFFT=2**14, label='Before filter')
f, s = plot_ASD(t, s_Livingston_filt, NFFT=2**14, label='After filter')
plt.ylabel("Strain noise, 1/$\sqrt{\mathrm{Hz}}$")
plt.legend()

fig.set_size_inches(13,4)
plt.show()
```

```python
def plot_filtered_wf(sH, sL, ylim=(-2e-21, 2e-21), xlim=(0,32) ):
    ## helper function to make plots of recovered filtered streams
    plt.plot(t, sH, label="Hanford" )
    plt.plot(t, sL, label="Livingston"  )
    plt.xlim(*xlim)
    plt.ylim(*ylim)
    plt.xlabel("Time (s)")
    plt.ylabel("Strain")
    plt.legend()
    fig.set_size_inches(14,3)

# plot the data over the full time range
fig=plt.figure()
plot_filtered_wf(s_Hanford_filt, s_Livingston_filt)
plt.show()
```

Note that the increased noise at the edges of the vector are similar "edge effects" to those that we previously applied the window function to attenuate. However, an additional burst of larger than normal strain is also visible near the center of the waveform. Zooming in, the pulse can be seen around 15.4 s into the waveform:

```python
## Zoom in t around the merger event
fig=plt.figure()
plot_filtered_wf(s_Hanford_filt, s_Livingston_filt, xlim=(15.1, 15.7))
plt.show()
```

```python
## Now very tight zoom to see time different
fig=plt.figure()
plot_filtered_wf(s_Hanford_filt, s_Livingston_filt, xlim=(15.4, 15.44))
plt.show()
```

When zoomed in to the maximum level, one can observe that the GW event arrives at Livingston before it arrives at Hanford. This is most clearly visible from the last two peaks in the waveform, which are offset by about one cycle between the two waves. The time delay for the wave to end at Hanford is ~7 ms after it ends at Livingston. This is consistent with the 3000 km distance between the sites, the finite speed of the GW at the speed of light $c$ (taking into account that the GW comes in a plane that is angled relative to the plane containing the Livingston and Hanford detectors).

Finally, we note if we zoom in on this portion of the waveform without filtering, the signal is hopelessly lost in the noise:

```python
## plot the unfiltered waveforms for comparison
fig=plt.figure()
plot_filtered_wf(s_Hanford, s_Livingston, xlim=(15.1, 15.7),ylim=(-2e-18,2e-18))
plt.show()
```
