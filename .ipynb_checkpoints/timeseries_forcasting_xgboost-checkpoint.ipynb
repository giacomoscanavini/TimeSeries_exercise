# Time Series Forecasting using XGBoost

```python
import numpy as np 
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
import xgboost as xgb
from xgboost import plot_importance, plot_tree
from sklearn.metrics import mean_squared_error, mean_absolute_error
plt.style.use('fivethirtyeight')
```

## Import Data Set
Import hourly power consumption data.

```python
pjme = pd.read_csv('data/PJME_hourly.csv', index_col=[0], parse_dates=[0])
```

```python
plt.figure(figsize=(20,5))
plt.title('Power Used')
pjme['PJME_MW'].plot.line()
plt.show()
```

```python
plt.figure(figsize=(20,5))
plt.title('Power Used')
pjme['PJME_MW'].resample('12H').mean().plot.line()
plt.show()
```

```python
plt.figure(figsize=(20,5))
plt.title('Power Used')
pjme['PJME_MW'].resample('10D').mean().plot.line()
plt.show()
```

# Train/Test Split
Split data before and after 2015, use data after 2015 as validation

```python
split_date = '01-Jan-2015'
pjme_train = pjme[pjme.index <= split_date].copy()
pjme_test = pjme[pjme.index > split_date].copy()
```

```python
plt.figure(figsize=(20,5))
plt.title('Power Used')
pjme_train['PJME_MW'].resample('5D').mean().plot.line(label='Train')
pjme_test['PJME_MW'].resample('5D').mean().plot.line(label='Test')
plt.legend()
plt.show()
```

# Create Time Series Features

```python
def create_features(df, label=None):
    """
    Creates time series features from datetime index value
    Information and methods can be found here: 
        https://pandas.pydata.org/docs/reference/api/pandas.Series.dt.html#
    """
    df['date'] = df.index
    df['hour'] = df['date'].dt.hour
    df['dayofweek'] = df['date'].dt.dayofweek # 0:Monday, 6:Sunday
    df['quarter'] = df['date'].dt.quarter
    df['month'] = df['date'].dt.month # 1:January, 12:December
    df['year'] = df['date'].dt.year
    df['dayofyear'] = df['date'].dt.dayofyear
    df['dayofmonth'] = df['date'].dt.day
    df['weekofyear'] = df['date'].dt.isocalendar().week.astype(int)
    
    X = df[['hour','dayofweek','quarter','month','year','dayofyear','dayofmonth','weekofyear']]
    if label:
        y = df[label]
        return X, y
    return X
```

```python
X_train, y_train = create_features(pjme_train, label='PJME_MW')
X_test, y_test = create_features(pjme_test, label='PJME_MW')
```

```python
X_train.describe()
```

# Create XGBoost Model

```python
reg = xgb.XGBRegressor(n_estimators=1000)
reg.fit(X_train, y_train,
        eval_set=[(X_train, y_train), (X_test, y_test)],
        early_stopping_rounds=50,
        verbose=False) # Change verbose to True if you want to see it train
```

## Feature Importances

```python
plot_importance(reg, height=0.7)
```

**Day of the year** seems to be the most important feature used in the node splitting, followed by **year** and **hour**

## Prediction on Test Sample

```python
pjme_test['Prediction'] = reg.predict(X_test)
pjme_all = pd.concat([pjme_test, pjme_train], sort=False)
```

```python
plt.figure(figsize=(20,5))
plt.title('Power Used')
pjme_train['PJME_MW'].resample('5D').mean().plot.line(label='Train')
pjme_test['PJME_MW'].resample('5D').mean().plot.line(label='Test')
pjme_test['Prediction'].resample('5D').mean().plot.line(label='Predicted')
plt.legend()
plt.show()
```

## Evaluation of Test Prediction
Metrics used: 
- MAE (mean absolute error) 
- MAPE (mean absolute percentage error)
- MSE (mean square error)

```python
mean_squared_error(y_true=pjme_test['PJME_MW'], y_pred=pjme_test['Prediction'])
```

```python
mean_absolute_error(y_true=pjme_test['PJME_MW'], y_pred=pjme_test['Prediction'])
```

```python
def mean_absolute_percentage_error(y_true, y_pred): 
    y_true, y_pred = np.array(y_true), np.array(y_pred)
    return np.mean(np.abs((y_pred-y_true) / y_true)) * 100
```

```python
mean_absolute_percentage_error(y_true=pjme_test['PJME_MW'], y_pred=pjme_test['Prediction'])
```

## Best and Worst predicted days

```python
pjme_test['error'] = pjme_test['PJME_MW'] - pjme_test['Prediction']
pjme_test['abs_error'] = np.abs(pjme_test['error'])
```

```python
# Mean over each hour
error_by_hour = pjme_test.groupby(['year','month','dayofmonth', 'hour'])[['PJME_MW','Prediction','error','abs_error']].mean()
error_by_hour.head(5)
```

```python
# Mean over each day
error_by_day = pjme_test.groupby(['year','month','dayofmonth'])[['PJME_MW','Prediction','error','abs_error']].mean()
error_by_day.head(5)
```

```python
# Mean over each month
error_by_month = pjme_test.groupby(['year','month'])[['PJME_MW','Prediction','error','abs_error']].mean()
error_by_month.head(5)
```

```python
# Best forecasted days
error_by_day.sort_values('abs_error', ascending=True).head(5)
```

```python
# Worst forecasted days
error_by_day.sort_values('abs_error', ascending=False).head(5)
```
