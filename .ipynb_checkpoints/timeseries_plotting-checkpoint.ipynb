# Time-series plotting

```python
import pandas as pd
from pandas.plotting import lag_plot
from pandas.plotting import autocorrelation_plot

import numpy as np
import matplotlib.pyplot as plt
#pd.set_option('max_columns', None)
```

## Import Data Set
Import data from an animal shelter

```python
shelter_outcomes = pd.read_csv("data/aac_shelter_outcomes.csv", parse_dates=['date_of_birth', 'datetime'])
shelter_outcomes = shelter_outcomes[['outcome_type', 'age_upon_outcome', 'datetime', 'animal_type', 'breed', 
                                     'color', 'sex_upon_outcome', 'date_of_birth']]
shelter_outcomes.head()
```

```python
plt.figure()
plt.title('Date of birth')
shelter_outcomes['date_of_birth'].value_counts().sort_values().plot.line()
plt.show()
```

- It looks like birth dates for the animals in the dataset peak at around 2015, but it's hard to tell for sure because the data is rather noisy.
- Resample birth date by year to see if the distribution peaks in 2015

```python
plt.figure()
plt.title('Date of birth')
shelter_outcomes['date_of_birth'].value_counts().resample('Y').sum().plot.line()
plt.show()
```

## Import data set
- Import stock data set
- Limit entries to symbol 'GOOG'
- Reset index on 'date'

```python
stocks = pd.read_csv("data/prices.csv", parse_dates=['date'])
stocks1 = stocks[stocks['symbol'] == "GOOG"].set_index('date')
stocks2 = stocks[stocks['symbol'] == "WLTW"].set_index('date')
stocks1.head(15)
```

### Visualize mean price of stock on year basis 'Y'
`pandas` is "datetime-aware". Usually the value of time-series data is exposed through this sort of grouping based on time periods.

```python
plt.figure(figsize=(12,5))
plt.subplot(121)
plt.title('Average volume')
stocks1['volume'].resample('Y').mean().plot.bar()
plt.subplot(122)
plt.title('Average open stock price [$]')
stocks1['open'].resample('Y').mean().plot.bar()
plt.show()
```

### Lag plot

A lag plot compares data points from each observation in the dataset against data points from a previous observation. So for example, data from December 21st will be compared with data from December 20th, which will in turn be compared with data from December 19th, and so on.

```python
plt.figure(figsize=(12,5))
plt.subplot(121)
plt.title(r'Volume at day $t$ and $t+1$')
lag_plot(stocks1['volume'].tail(250))
plt.subplot(122)
plt.title(r'Average close stock price at day $t$ and $t+1$ [$]')
lag_plot(stocks1['close'].tail(250))
plt.show()
```

- There seem to exist correlation between the volume of a stock on a specific day and the volume of the same stock the previous day. The correlation seems stronger (as expected) for the price of the stock. 
- Time-series data tends to exhibit a behavior called **periodicity**: rises and peaks in the data that are correlated with time. 
- Lag plots are extremely useful because they are a simple way of checking datasets for this kind of periodicity.

### Autocorrelation plot
- The autocorrelation plot is a multivariate summarization-type plot that lets you check *every* periodicity at the same time. It does this by computing the correlation between a series with itself at a previous time in the past (lag). This is computed for all values of lag. 
- Autocorrelation helps in identifying repeating patterns, such as the presence of a periodic signal obscured by noise, or identifying the missing fundamental frequency in a signal implied by its harmonic frequencies. 
- In an autocorrelation plot the lag is on the x-axis and the autocorrelation score is on the y-axis. The farther away the autocorrelation is from 0, the greater the influence that records that far away from each other exert on one another.

```python
plt.figure(figsize=(12,5))
plt.subplot(121)
plt.title(r'Autocorrealtion on volume')
autocorrelation_plot(stocks1['volume'])
plt.subplot(122)
plt.title(r'Autocorrelation on stock price')
autocorrelation_plot(stocks1['close'])
plt.show()
```

It seems like the volume of trading activity is weakly descendingly correlated with trading volume from the year prior. There aren't any significant non-random peaks in the dataset, so this is good evidence that there isn't much of a time-series pattern to the volume of trade activity over time.
